```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
def create_3d_gaussplot(scale_x, scale_y, figsize=(12, 10), color='#598f70', alpha=1.0):
    """ Used to create a 3D Gaussian surface plot."""
    
    x, y = np.mgrid[-1.5:1.5:50j, -1.5:1.5:50j]
    xy = np.column_stack([x.flat, y.flat])
    mu = np.array([0.0, 0.0])
    sigma = np.array([scale_x, scale_y])
    covariance = np.diag(sigma**2)
    z = multivariate_normal.pdf(xy, mean=mu, cov=covariance)
    z = z.reshape(x.shape)

    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x,y,z, color=color, alpha=alpha)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_wireframe(x,y,z, color=color, alpha=alpha)
    ax.axis('off')
    fig.tight_layout()
    return fig
```

```python
scales = [
    (1.25, 1.25),
    (0.9, 0.9),
    (0.65, 0.65),
    (0.4, 0.4)
]

for i, scale in enumerate(scales):
    fig = create_3d_gaussplot(scale[0], scale[1], figsize=(10, 10), color='#852626', alpha=0.75)
    fig.savefig(f'../figures/3d_gaussian_{i}.png', dpi=300, bbox_inches='tight')
```

```python
fig = create_3d_gaussplot(10.0, 10.0, figsize=(10, 10), color='#c47878', alpha=0.75)
fig.savefig('../figures/3d_gaussian_prior.png', dpi=300, bbox_inches='tight')
```
